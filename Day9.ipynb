```python
def parse_file(file_name='input.dat'):
  with open(file_name) as f:
    lines = f.readlines()
  
  motions = []
  for line in lines:
    line = line.replace('\n', '')
    s = line.split(' ')
    dist = int(s[1])
    if s[0] == 'R':
      motions.append((dist, 0))
    elif s[0] == 'L':
      motions.append((-dist, 0))
    elif s[0] == 'U':
      motions.append((0, dist))
    else:
      motions.append((0, -dist))

  return motions
```

```python
motions = parse_file()
print(motions[0])
print(motions[-1])
```

```python
def get_tail_points(motions):
  visited = set([(0,0)])

  head = (0,0)
  tail = (0,0)

  for move in motions:
    # move left
    for i in range(move[0], 0):
      # print("L ", abs(move[0]))
      temp_head = (head[0] -1, head[1])
      if should_move(temp_head, tail):
        tail = head
        
      # else:
      #   print("no move.  Head: ", temp_head, ", tail: ", tail)
      head = temp_head
      # print("Head: ", temp_head, ", tail: ", tail)
      visited.add(tail)

    # move right
    for i in range(0, move[0]):
      # print("R ", abs(move[0]))
      temp_head = (head[0] + 1, head[1])
      if should_move(temp_head, tail):
        tail = head
        
      # else:
      #   print("no move.  Head: ", temp_head, ", tail: ", tail)
      head = temp_head
      # print("Head: ", temp_head, ", tail: ", tail)
      visited.add(tail)
    # move up
    for i in range(0,move[1]):
      # print("U ", abs(move[1]))
      temp_head = (head[0], head[1] + 1)
      if should_move(temp_head, tail):
        tail = head
      # else:
      #   print("no move.  Head: ", temp_head, ", tail: ", tail)
      head = temp_head
      # print("Head: ", temp_head, ", tail: ", tail)
      visited.add(tail)
    # move down
    for i in range(move[1], 0):
      # print("D ", abs(move[1]))
      temp_head = (head[0], head[1] - 1)
      if should_move(temp_head, tail):
        tail = head
        # print("move Head: ", temp_head, ", tail: ", tail)
      # else:
      #   print("no move.  Head: ", temp_head, ", tail: ", tail)
      head = temp_head
      # print("Head: ", temp_head, ", tail: ", tail)
      visited.add(tail)
    
    # print('head: ', head, ', tail:', tail)
    

  return visited
```

```python
def should_move(head, tail):
  return (abs(head[0] - tail[0]) > 1 ) or (abs(head[1] - tail[1]) > 1)
```

```python
visited = get_tail_points(motions)
```

```python
def get_move(head, tail):
  horizontal = head[0] - tail[0]
  vertical = head[1] - tail[1]
  x_move = 0
  y_move = 0
  if horizontal > 1:
    x_move = 1
    y_move = max(min(vertical, 1), -1)
  elif horizontal < -1:
    x_move = -1
    y_move = max(min(vertical, 1), -1)
  if vertical > 1:
    x_move = max(min(horizontal, 1), -1)
    y_move = 1
  elif vertical < -1:
    x_move = max(min(horizontal, 1), -1)
    y_move = -1
  move = (tail[0] + x_move, tail[1] + y_move)
  # print("head:", head, ", tail:", tail, ", move:", (x_move, y_move) )
  return (tail[0] + x_move, tail[1] + y_move)
```
